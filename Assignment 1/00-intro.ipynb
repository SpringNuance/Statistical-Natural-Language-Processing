### ENCODE TO PIG LATIN

**Pig Latin** aka **Igpay Atinlay** is a "secret language" (or a word game) that encodes words in English by
* **1** moving the first consonant or sequence of consonants to the end of the word (pig -> igp)
* **2** adding "ay" to the end of the word (igp -> igpay)
* **3** just adding "ay" to the end of the word if the word starts with a vowel (owl -> owlay)

Let's assume that the letters corresponding to English vowels in the beginning of the words are **A**, **E**, **I**, **O**, **U**, and **W**. Consider all other letters as consonants here (this assumption doesn't always hold in real life: **H** in "honest" produces a vowel). For the sake of the assignment, let's pretend that only lowercase letters exist in the world and there are no empty strings.

Complete the `pig_latin()` function with you code after '**# YOUR CODE HERE**' comment.

Don't forget to delete `raise NotImplementedError()` when you're done with your answer!

This task introduces autograded code answers. There are some visible tests and some hidden tests for your answer. Hidden tests help us make sure your functions actually work. Don't rely on visible tests alone in further assignments, they might not cover all the corner cases. For example, in this task, there are no visible tests for words starting with vowels. So we encourage you to write your own tests, but we will try to keep visible tests as helpful as possible.

```python
def pig_latin(word):
    """This function takes in a word and outputs its Pig Latin translation
    
    Parameters
    ----------
    word - str
        a word to translate to Pig Latin
    
    Returns
    -------
    pig_word - str
        a word translated into Pig Latin as a string
    
    """
    # YOUR CODE HERE
    
    # Check if the first letter is a vowel
    if word[0] in ["a","i","u","e","o","w"]:
        pig_word = word + 'ay'
    else:
        # Find the position of the first vowel
        for i, letter in enumerate(word):
            if letter in 'aeiou':
                break
            else:
                # No vowel found, consider entire word as consonant
                i = len(word)

        # Move the first consonant or consonant cluster to the end and add 'ay'
        pig_word = word[i:] + word[:i] + 'ay'

    # raise NotImplementedError()
    return pig_word
```

```python
assert(pig_latin("sort") == "ortsay")
assert(pig_latin("smile") == "ilesmay")
assert(pig_latin("snlp") == "snlpay")
```
